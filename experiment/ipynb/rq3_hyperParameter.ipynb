```python
import pandas as pd

def readCSV_base_hp(file_path):
  column_names = ['index', 'eps', 
                  'base_type', 'base_time', 'base_visit', 'base_status', 'base_lb',
                  'lam_0_c_0_type', 'lam_0_c_0_time', 'lam_0_c_0_visit', 'lam_0_c_0_status', 'lam_0_c_0_lb', 
                  'lam_0_c_02_type', 'lam_0_c_02_time', 'lam_0_c_02_visit', 'lam_0_c_02_status', 'lam_0_c_02_lb', 
                  'lam_0_c_1_type', 'lam_0_c_1_time', 'lam_0_c_1_visit', 'lam_0_c_1_status', 'lam_0_c_1_lb', 
                  'lam_05_c_0_type', 'lam_05_c_0_time', 'lam_05_c_0_visit', 'lam_05_c_0_status', 'lam_05_c_0_lb', 
                  'lam_05_c_02_type', 'lam_05_c_02_time', 'lam_05_c_02_visit', 'lam_05_c_02_status', 'lam_05_c_02_lb', 
                  'lam_05_c_1_type', 'lam_05_c_1_time', 'lam_05_c_1_visit', 'lam_05_c_1_status', 'lam_05_c_1_lb', 
                  'lam_1_c_0_type', 'lam_1_c_0_time', 'lam_1_c_0_visit', 'lam_1_c_0_status', 'lam_1_c_0_lb', 
                  'lam_1_c_02_type', 'lam_1_c_02_time', 'lam_1_c_02_visit', 'lam_1_c_02_status', 'lam_1_c_02_lb', 
                  'lam_1_c_1_type', 'lam_1_c_1_time', 'lam_1_c_1_visit', 'lam_1_c_1_status', 'lam_1_c_1_lb', 
                  'lam_0_c_04_type', 'lam_0_c_04_time', 'lam_0_c_04_visit', 'lam_0_c_04_status', 'lam_0_c_04_lb', 
                  'lam_0_c_06_type', 'lam_0_c_06_time', 'lam_0_c_06_visit', 'lam_0_c_06_status', 'lam_0_c_06_lb', 
                  'lam_0_c_08_type', 'lam_0_c_08_time', 'lam_0_c_08_visit', 'lam_0_c_08_status', 'lam_0_c_08_lb',
                  'lam_05_c_04_type', 'lam_05_c_04_time', 'lam_05_c_04_visit', 'lam_05_c_04_status', 'lam_05_c_04_lb', 
                  'lam_05_c_06_type', 'lam_05_c_06_time', 'lam_05_c_06_visit', 'lam_05_c_06_status', 'lam_05_c_06_lb', 
                  'lam_05_c_08_type', 'lam_05_c_08_time', 'lam_05_c_08_visit', 'lam_05_c_08_status', 'lam_05_c_08_lb', 
                  'lam_1_c_04_type', 'lam_1_c_04_time', 'lam_1_c_04_visit', 'lam_1_c_04_status', 'lam_1_c_04_lb', 
                  'lam_1_c_06_type', 'lam_1_c_06_time', 'lam_1_c_06_visit', 'lam_1_c_06_status', 'lam_1_c_06_lb', 
                  'lam_1_c_08_type', 'lam_1_c_08_time', 'lam_1_c_08_visit', 'lam_1_c_08_status', 'lam_1_c_08_lb']
  df = pd.read_csv(file_path, sep=",", header=None, names=column_names)
  return df
```

## speedup ratio (base/hp)

```python
def speedUpRatio(filePath):
  all_df = readCSV_base_hp(filePath)
  time_df = all_df[[col for col in all_df.columns if 'time' in col]]
  
  return [
    [
      (time_df.base_time/time_df.lam_0_c_0_time).mean(),
      (time_df.base_time/time_df.lam_0_c_02_time).mean(),
      (time_df.base_time/time_df.lam_0_c_04_time).mean(),
      (time_df.base_time/time_df.lam_0_c_06_time).mean(),
      (time_df.base_time/time_df.lam_0_c_08_time).mean(),
      (time_df.base_time/time_df.lam_0_c_1_time).mean()
    ],[
      (time_df.base_time/time_df.lam_05_c_0_time).mean(),
      (time_df.base_time/time_df.lam_05_c_02_time).mean(),
      (time_df.base_time/time_df.lam_05_c_04_time).mean(),
      (time_df.base_time/time_df.lam_05_c_06_time).mean(),
      (time_df.base_time/time_df.lam_05_c_08_time).mean(),
      (time_df.base_time/time_df.lam_05_c_1_time).mean()
    ],[
      (time_df.base_time/time_df.lam_1_c_0_time).mean(),
      (time_df.base_time/time_df.lam_1_c_02_time).mean(),
      (time_df.base_time/time_df.lam_1_c_04_time).mean(),
      (time_df.base_time/time_df.lam_1_c_06_time).mean(),
      (time_df.base_time/time_df.lam_1_c_08_time).mean(),
      (time_df.base_time/time_df.lam_1_c_1_time).mean()
    ]
  ]
```

## time average

```python
def timeAverage(filePath):
  all_df = readCSV_base_hp(filePath)
  time_df = all_df[[col for col in all_df.columns if 'time' in col]]
  
  return [
    [
      (time_df.lam_0_c_0_time).mean(),
      (time_df.lam_0_c_02_time).mean(),
      (time_df.lam_0_c_04_time).mean(),
      (time_df.lam_0_c_06_time).mean(),
      (time_df.lam_0_c_08_time).mean(),
      (time_df.lam_0_c_1_time).mean()
    ],[
      (time_df.lam_05_c_0_time).mean(),
      (time_df.lam_05_c_02_time).mean(),
      (time_df.lam_05_c_04_time).mean(),
      (time_df.lam_05_c_06_time).mean(),
      (time_df.lam_05_c_08_time).mean(),
      (time_df.lam_05_c_1_time).mean()
    ],[
      (time_df.lam_1_c_0_time).mean(),
      (time_df.lam_1_c_02_time).mean(),
      (time_df.lam_1_c_04_time).mean(),
      (time_df.lam_1_c_06_time).mean(),
      (time_df.lam_1_c_08_time).mean(),
      (time_df.lam_1_c_1_time).mean()
    ]
  ]
```

## solved

```python
def solvedNum(filePath):
  all_df = readCSV_base_hp(filePath)
  
  return [
    [
      all_df[all_df.lam_0_c_0_status != 'unknown'].shape[0],
      all_df[all_df.lam_0_c_02_status != 'unknown'].shape[0],
      all_df[all_df.lam_0_c_04_status != 'unknown'].shape[0],
      all_df[all_df.lam_0_c_06_status != 'unknown'].shape[0],
      all_df[all_df.lam_0_c_08_status != 'unknown'].shape[0],
      all_df[all_df.lam_0_c_1_status != 'unknown'].shape[0],
    ],[
      all_df[all_df.lam_05_c_0_status != 'unknown'].shape[0],
      all_df[all_df.lam_05_c_02_status != 'unknown'].shape[0],
      all_df[all_df.lam_05_c_04_status != 'unknown'].shape[0],
      all_df[all_df.lam_05_c_06_status != 'unknown'].shape[0],
      all_df[all_df.lam_05_c_08_status != 'unknown'].shape[0],
      all_df[all_df.lam_05_c_1_status != 'unknown'].shape[0],
    ],[
      all_df[all_df.lam_1_c_0_status != 'unknown'].shape[0],
      all_df[all_df.lam_1_c_02_status != 'unknown'].shape[0],
      all_df[all_df.lam_1_c_04_status != 'unknown'].shape[0],
      all_df[all_df.lam_1_c_06_status != 'unknown'].shape[0],
      all_df[all_df.lam_1_c_08_status != 'unknown'].shape[0],
      all_df[all_df.lam_1_c_1_status != 'unknown'].shape[0],
    ]
  ]
```

## make heatmap

```python
import matplotlib.pyplot as plt
import numpy as np

def makeHeatMap_speedup(filePath):

    data = speedUpRatio(filePath)

    # label
    lambdas = [0, 0.5, 1]  
    c_values = [0, 0.2, 0.4, 0.6, 0.8, 1]  

    fig, ax = plt.subplots(figsize=(6, 2))

    # heatmap
    heatmap = ax.imshow(data, cmap='YlGnBu', aspect='auto')

    # x label
    ax.set_xticks(np.arange(len(c_values)))
    ax.set_xticklabels([f'c: {c:.1f}' for c in c_values], fontsize=14)

    # y label
    ax.set_yticks(np.arange(len(lambdas)))
    ax.set_yticklabels([f'λ: {l:.1f}' for l in lambdas], fontsize=14)

    # put data
    for i in range(len(lambdas)):
        for j in range(len(c_values)):
            ax.text(j, i, f'{data[i][j]:.2f}', ha='center', va='center', 
                    color='black' if data[i][j] < 1.05 else 'white', fontsize=15)

    # color bar
    # cbar = plt.colorbar(heatmap, ax=ax, orientation='horizontal', pad=0.2)
    # cbar.set_label('Speedup', fontsize=14)
    # cbar.ax.tick_params(labelsize=12)

    # title
    # plt.title('Speedup(baseline/ABONN)', fontsize=16)

    # show and save
    plt.tight_layout()
    plt.savefig('../pdf/heatmap/heatmap_speedup.pdf', bbox_inches='tight')
    plt.show()
```
