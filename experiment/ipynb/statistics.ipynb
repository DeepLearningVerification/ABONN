```python
import pandas as pd
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)

def readCSV(file_path):
  column_names = ['index', 'eps', 
                  'abonn_type', 'abonn_time', 'abonn_visit', 'abonn_status', 'abonn_lb', 
                  'base_type', 'base_time', 'base_visit', 'base_status', 'base_lb', 
                  'ab_type', 'ab_status', 'ab_time']
  df = pd.read_csv(file_path, sep=",", header=None, names=column_names)
  return df
```

```python
def speedUp(fileName):
  # fileName_list = [
  #   'mnistl2',
  #   'mnistl4',
  #   'base',
  #   'deep',
  #   'wide',
  # ]
  filePath = f'../csv/{fileName}.csv'
  df = readCSV(filePath)
  filtered_df = df[df.base_status == df.abonn_status]
  filtered_df['speedUp'] = filtered_df['base_time'] / filtered_df['abonn_time']
  print(fileName)
  show_df = filtered_df[[col for col in filtered_df.columns if col.endswith('time') | col.endswith('status') | col.endswith('speedUp')]]
  return show_df  
```

```python
speedUp('mnistl2')
```

```python
speedUp('mnistl4')
```
